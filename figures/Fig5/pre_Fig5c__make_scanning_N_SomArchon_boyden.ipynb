```python
# make sliding-window of 50 replacements with Ns
window_size = 50
seqs = []
window_start_idxs = []
window_end_idxs = []
plasmid = np.array(list(plasmid))
for idx in range(len(plasmid) - window_size + 1):
    mutated = np.array(plasmid)
    mutated[idx:idx+window_size] = 'N'
    seqs.append(''.join(mutated.tolist()))
    window_start_idxs.append(idx)
    window_end_idxs.append(idx + window_size)
```

```python
# now do a sliding window of 10 Ns
window_size = 10
for idx in range(len(plasmid) - window_size + 1):
    mutated = np.array(plasmid)
    mutated[idx:idx+window_size] = 'N'
    seqs.append(''.join(mutated.tolist()))
    window_start_idxs.append(idx)
    window_end_idxs.append(idx + window_size)
```
